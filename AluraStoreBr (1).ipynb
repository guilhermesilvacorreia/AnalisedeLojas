# 2. Vendas por Categoria


```python
 ## estudo das vendas por categoria
def contar_categorias(dados_loja):
  contagem= {}
  for categoria in dados_loja['Categoria do Produto']:
    if categoria in contagem:
      contagem[categoria] += 1
    else:
      contagem[categoria] = 1
  return dict(sorted(contagem.items(), key=lambda item: item[1], reverse=True))

print(contar_categorias(loja_1))
print(contar_categorias(dados_loja2))
print(contar_categorias(dados_loja3))
print(contar_categorias(dados_loja4))
```

```python
 ### Criar dicionário principal com todas as lojas
resultados ={
    'Loja_1': contar_categorias(loja_1),
    'Loja_2': contar_categorias(dados_loja2),
    'Loja_3': contar_categorias(dados_loja3),
    'Loja_4': contar_categorias(dados_loja4)
}
# Mostrar categorias mais populares por loja
for loja, categorias in resultados.items():
    print(f"Categorias mais populares na {loja}:")
for categoria, quantidade in categorias.items():
    print(f"{categoria}: {quantidade} vendas")
mais_popular = next(iter(categorias))
print(f"A categoria mais popular é: {mais_popular}")

# Função para contar categorias
def contar_categorias_total(*dados_lojas):
    contagem_total = {}
    for dados_loja in dados_lojas:
        for categoria in dados_loja['Categoria do Produto']:
            if categoria in contagem_total:
                contagem_total[categoria] += 1
            else:
                contagem_total[categoria] = 1
    return dict(sorted(contagem_total.items(), key=lambda item: item[1], reverse=True))

# Contar categorias considerando todas as lojas
categorias_mais_vendidas = contar_categorias_total(loja_1, dados_loja2, dados_loja3, dados_loja4)

# Exibir o resultado
print("Categorias mais vendidas (somando todas as lojas):")
for categoria, quantidade in categorias_mais_vendidas.items():
    print(f"• {categoria}: {quantidade} vendas")

# Determinar a categoria mais vendida
categoria_mais_vendida = max(categorias_mais_vendidas, key=categorias_mais_vendidas.get)
print(f"A categoria mais vendida é: {categoria_mais_vendida}")



```

# 4. Produtos Mais e Menos Vendidos

```python
## Contando Frequência dos produtos
def contar_produtos(dados_loja):
    contagem = {}
    for produto in dados_loja['Produto']:
        contagem[produto] = contagem.get(produto, 0) + 1
    return contagem

# Função para obter os produtos mais e menos vendidos a partir de uma contagem
def produtos_extremos(contagem):
    if not contagem:
        return None, None
    max_vendas = max(contagem.values())
    min_vendas = min(contagem.values())

    # Lista com produtos com vendas igual ao máximo ou mínimo
    mais_vendidos = [produto for produto, qtd in contagem.items() if qtd == max_vendas]

    return mais_vendidos, [produto for produto, qtd in contagem.items() if qtd == min_vendas]
```

```python
 # Aplicando para cada loja:

# Loja 1
contagem_loja1 = contar_produtos(loja_1)
mv_loja1, lv_loja1 = produtos_extremos(contagem_loja1)
print("Loja 1:")
print("  Produtos mais vendidos:", mv_loja1, "com", contagem_loja1[mv_loja1[0]], "venda(s).")
print("  Produtos menos vendidos:", lv_loja1, "com", contagem_loja1[lv_loja1[0]], "venda(s).")

# Loja 2
contagem_loja2 = contar_produtos(dados_loja2)
mv_loja2, lv_loja2 = produtos_extremos(contagem_loja2)
print("\nLoja 2:")
print("  Produtos mais vendidos:", mv_loja2, "com", contagem_loja2[mv_loja2[0]], "venda(s).")
print("  Produtos menos vendidos:", lv_loja2, "com", contagem_loja2[lv_loja2[0]], "venda(s).")

# Loja 3
contagem_loja3 = contar_produtos(dados_loja3)
mv_loja3, lv_loja3 = produtos_extremos(contagem_loja3)
print("\nLoja 3:")
print("  Produtos mais vendidos:", mv_loja3, "com", contagem_loja3[mv_loja3[0]], "venda(s).")
print("  Produtos menos vendidos:", lv_loja3, "com", contagem_loja3[lv_loja3[0]], "venda(s).")

# Loja 4
contagem_loja4 = contar_produtos(dados_loja4)
mv_loja4, lv_loja4 = produtos_extremos(contagem_loja4)
print("\nLoja 4:")
print("  Produtos mais vendidos:", mv_loja4, "com", contagem_loja4[mv_loja4[0]], "venda(s).")
print("  Produtos menos vendidos:", lv_loja4, "com", contagem_loja4[lv_loja4[0]], "venda(s).")


```

## Introdução

Este relatório detalha a análise de desempenho de quatro lojas distintas, com o objetivo de fornecer subsídios para uma decisão estratégica. A avaliação baseou-se em indicadores-chave, incluindo faturamento total, distribuição de categorias de produtos, avaliações médias de clientes e custo médio de frete. A visualização de dados através de gráficos auxiliou na identificação de tendências e particularidades de cada estabelecimento.

## Análise de Resultados

**1. Faturamento:** Observou-se uma variação significativa no volume de receita entre as lojas. A Loja 3 destacou-se com o maior faturamento (aproximadamente R$ 3.278), seguida pela Loja 2 (R$ 2.810). As Lojas 1 e 4 apresentaram faturamentos idênticos e mais modestos (R$ 2.372 cada). O desempenho superior da Loja 3 sugere maior penetração de mercado ou estratégias de vendas mais agressivas.

**2. Satisfação do Cliente:** As avaliações médias revelaram um cenário inverso ao do faturamento. As Lojas 1 e 4 lideram em satisfação, com nota média de 4.2, indicando uma experiência de compra positiva. Em contrapartida, a Loja 3, apesar do alto volume de vendas, registrou a menor média de satisfação (3.4), seguida pela Loja 2 (3.6). Este dado aponta para potenciais fragilidades na experiência do cliente ou na qualidade percebida na Loja 3.

**3. Custo de Frete:** Os custos logísticos, representados pelo frete médio, também variaram. As Lojas 1 e 4 apresentaram os custos mais competitivos (aproximadamente R$ 23,50). A Loja 2 posicionou-se em um nível intermediário (R$ 28,60), enquanto a Loja 3 registrou o frete médio mais elevado (R$ 33,60). Custos de frete mais baixos tendem a aumentar a atratividade da loja e a conversão de vendas.

**4. Mix de Produtos:** A análise das categorias mostrou que "móveis" é um segmento relevante em todas as lojas, especialmente na Loja 2, onde domina as vendas. As Lojas 1 e 4 demonstraram uma distribuição mais equilibrada entre diversas categorias (eletrodomésticos, móveis, esporte, livros, eletrônicos), sugerindo maior diversificação. A Loja 3 também mostra vendas concentradas em móveis, mas com presença de outras categorias como eletrodomésticos e instrumentos musicais.

## Conclusão e Recomendação

A análise integrada dos indicadores revela um quadro complexo. A Loja 3, embora líder em faturamento, apresenta desafios críticos relacionados à baixa satisfação do cliente e ao alto custo de frete, fatores que podem comprometer a sustentabilidade do negócio a longo prazo.

Por outro lado, a Loja 1 (e, similarmente, a Loja 4) demonstra um equilíbrio mais vantajoso. Seu faturamento consistente é complementado por altos níveis de satisfação do cliente e custos de frete competitivos. Essa combinação favorece a fidelização de clientes e a construção de uma reputação sólida, além de otimizar a competitividade no mercado.

Diante do exposto, **recomenda-se a Loja 1 como a opção mais estratégica**. A escolha se fundamenta na sua capacidade de aliar desempenho de vendas satisfatório a uma experiência de cliente superior e custos logísticos eficientes. Tal equilíbrio configura uma base mais robusta para crescimento contínuo e rentabilidade sustentável.
